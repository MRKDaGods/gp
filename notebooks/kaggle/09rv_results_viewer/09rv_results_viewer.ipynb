# 09r + 09s Results Viewer

```python
import json, pathlib, os

print("=== /kaggle/input/ directory structure ===")
input_root = pathlib.Path("/kaggle/input")
all_files = sorted(input_root.rglob("*"))
print(f"Total files found: {len(all_files)}")
for f in all_files[:50]:
    size = f.stat().st_size if f.is_file() else 0
    print(f"  {f}  ({size} bytes)" if f.is_file() else f"  {f}/")

print("\n=== Searching for summary JSONs ===")
for f in all_files:
    if f.is_file() and "summary" in f.name.lower() and f.suffix == ".json":
        print(f"Found: {f}")
        try:
            d = json.loads(f.read_text())
            print(json.dumps(d, indent=2))
        except Exception as e:
            print(f"  Error reading: {e}")

print("\n=== Searching for run_log.txt ===")
for f in all_files:
    if f.is_file() and f.name == "run_log.txt":
        print(f"Found: {f}")
        lines = f.read_text(errors="replace").splitlines()
        print(f"  Total lines: {len(lines)}")
        print("  Last 100 lines:")
        for line in lines[-100:]:
            print("  ", line)
```
